```python
# Compare Step Count and PAM Score Between Two Devices
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load both CSVs
device1 = pd.read_csv("multiple_A.csv", parse_dates=["Timestamp"])
device2 = pd.read_csv("multiple_B.csv", parse_dates=["Timestamp"])

# Merge on Timestamp (assuming timestamps are aligned already)
merged = pd.merge(device1, device2, on="Timestamp", suffixes=('_dev1', '_dev2'))

# Calculate differences
merged["Steps_diff"] = merged["Steps_dev1"] - merged["Steps_dev2"]
merged["PAM_diff"] = merged["PAM Score_dev1"] - merged["PAM Score_dev2"]

# --- Plot Step Comparison ---
plt.figure(figsize=(14, 6))
plt.plot(merged["Timestamp"], merged["Steps_dev1"], label="Device 1 Steps")
plt.plot(merged["Timestamp"], merged["Steps_dev2"], label="Device 2 Steps", linestyle='--')
plt.title("Step Comparison")
plt.xlabel("Time")
plt.ylabel("Steps")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# --- Plot PAM Score Comparison ---
plt.figure(figsize=(14, 6))
plt.plot(merged["Timestamp"], merged["PAM Score_dev1"], label="Device 1 PAM Score")
plt.plot(merged["Timestamp"], merged["PAM Score_dev2"], label="Device 2 PAM Score", linestyle='--')
plt.title("PAM Score Comparison")
plt.xlabel("Time")
plt.ylabel("PAM Score")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# --- Show scores of each file and difference ---
comparison_table = pd.DataFrame({
    'Timestamp': merged['Timestamp'],
    'Steps_Device1': merged['Steps_dev1'],
    'Steps_Device2': merged['Steps_dev2'],
    'Steps_Diff': merged['Steps_diff'],
    'PAM_Score_Device1': merged['PAM Score_dev1'],
    'PAM_Score_Device2': merged['PAM Score_dev2'],
    'PAM_Score_Diff': merged['PAM_diff']
})

# Display the comparison table
comparison_table.head(25)  # You can remove the head() to see the whole table
```
